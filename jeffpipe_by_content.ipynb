<h3><center><font color='GREEN'>PROGRAM TO SEPARATE SHAPEFILES FOR DIFFERENT MATERIAL THEY CARRY IN JEFFERSON COUNTY</font></center></h3>

<h4><font color='GREEN'> Loading Libraries </font> </h4>

```python
#Importing GeoPandas library to import shapefile
import geopandas as gpd
a = gpd.read_file('D:\\Spring23\\RA\\pipeline245.zip')
```

```python
a.head()
```

```python
#extracting the unique pipeline carrying material present in 'CMDTY_DESC' column as an array
b=a['CMDTY_DESC'].unique()
```

<h4><font color='GREEN'> FUNCTION TO SEPERATE SHAPEFILES BASED ON PIPE CONTENT </font> </h4>

```python
def shp_pipe_carry(c,i):
    
    mod_shp=a.loc[a['CMDTY_DESC']==c]
    
    return mod_shp
```

```python
i=0 # initializing an integer i

#creating a for loop to group and save pipeline carrying same materials in to seperate shapefiles
for i in range(0, len(b)):
    shp_new=shp_pipe_carry (b[i],i)
    shp_new.to_file('D:\\Spring23\\RA\\pipeline245\\by_content\\jeff_'+b[i]+'.shp')  
```
